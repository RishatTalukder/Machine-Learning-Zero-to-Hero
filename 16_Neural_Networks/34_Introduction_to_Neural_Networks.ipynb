# Introduction

`Neural Networks` are a set of `algorithms`, modeled loosely after the `human brain`, that are designed to `recognize patterns`. They `interpret sensory data` through a kind of `machine perception`, `labeling`, or `clustering` of raw input. The patterns they recognize are `numerical` and contained in `vectors`, into which all real-world data (images, sound, text, etc.) must be `translated`.

> Without any of the weird words, Neural Networks are a `set of algorithms` that help us to `model complex patterns` and `prediction problems` that are `non-linear` in nature.


## Perceptron Model

To fully unserstand `neural networks`, we need to understand the `Perceptron Model`.

A `perceptron` is the most basic `building block` of an `artificial neural network`. Invented by Frank Rosenblatt in 1957. 

It is a simplified mathematical model of a single biological neuron. 

You can think of a `Perceptron` as a `single neuron` of a `Human Brain`. It takes `input`, processes it, and provides `output`.

If you think of a `Perceptron` as a `single neuron`, this is how it looks like:

```mermaid

graph LR
    A[Input] --> B[Processing Unit]
    B --> C[Output]

```

It is the `simplest form` of a `neural network`. 

Amazingly, even back almost `70 years ago`, `Frank Rosenblatt` was able to see the potential of `neural networks` and stated that...

> "The `Perceptron` may eventually be able to `learn`, `make decisions`, and `translate languages`."

Which is exactly what we are doing today with `Deep Learning` and `Neural Networks` and `Machine Learning` and `Artificial Intelligence` and in the base of all these, there is the `Perceptrons`.

And this is why it is important to understand the `Perceptron` model.

But it was not until the `1980s` that `Neural Networks` started to become popular again. Because in `1969`, `Marvin Minsky` and `Seymour Papert` published a book called `Perceptrons` which showed the `limitations` of `Perceptrons`. They showed that `Perceptrons` were `limited` to `linearly separable problems`. The book also suggested that `Perceptrons` were `severely limited` in what they could learn.

This Statement by `Marvin Minsky` and `Seymour Papert` led to the `first AI winter` where `funding` and `interest` in `Neural Networks` and `AI` dried up.

SO, back in the `1970s` There was not enough `computational power` to `train` Neural Networks and Perceptrons were limited to linearly separable problems.

But in the `1980s`, `Neural Networks` started to become popular again with the `discovery` of the `backpropagation algorithm` which allowed `Neural Networks` to learn `non-linear functions`.

### Perceptrons

Now again think of the `screen` you are looking at right now. It is made up of `pixels`. Each `pixel` has a `color` and `intensity`. If you think of a `pixel` as a `feature`, then a `Perceptron` can be used to `classify` the `pixel` as `black` or `white`. SO, by analyzing the the screen we can see, almost always, we are taking `multiple inputs` and `processing` them to get `output`. 

So, now suppose we have `two inputs` (X1, X2) and we want to `classify` them as `black` or `white` which we will represent as `y`. We can represent this as:



```mermaid

graph LR
    A[X1] --> B[Perceptron]
    C[X2] --> B
    B --> D[y]

```

Now, the `Perceptron` will take `input` from `X1` and `X2`, process it, and provide `output` as `y`.

We can represent the `Perceptron` as a `mathematical model` or a `function`. So, the final `perceptron` will look like this:

```mermaid

graph LR
    A[X1] --> B["Perceptron or f(x)"]
    C[X2] --> B
    B --> D[y]

```

So, now if I say that the `f(x)` is a `sum` then,

$$ f(x) = x_1 + x_2 $$

Simple as that. This is the most simplistic form of a `Perceptron`. It takes `input`, `processes` it, and provides `output`.

Now Realistically, a we want to be able to adjust some `parameters` of the `Perceptron` to make it more `flexible` and `powerful`. In the above model there is no flexibility for the Perceptron to learn from the data. So, what we can do is add a `weight` to the `input` which will help the `Perceptron` to `learn` from the `data` more `efficiently`. So, we will add the `weights` to the `input` as a `multiplication` factor. So, the `Perceptron` will look like this:

```mermaid

graph LR
    A[X1] --> |w1| B(("Perceptron or f(x)"))
    C[X2] --> |w2| B
    B --> D[y]

```

So, the `Perceptron` will take `input` from `X1` and `X2`, `multiply` it with `weights` `w1` and `w2`, `process` it, and provide `output` as `y`.

$$ y = w_1x_1 + w_2x_2 $$

So, now we can `adjust` the `weights` to make the Perceptron more accurate and select the `weights` that `minimize` the `error`. This will give us more `flexibility` to the `Perceptron` to `learn` from the `data`.

> We can think of the `weights` as `parameters` of the Perceptron that we can adjust to make the Perceptron more accurate.

Now here's a brain teaser for you. What if the actual `input` is `X1` or `X2` is `0`? The `weights` will not have any effect on the `output`. How do we fix this? 

The answer is to add a `bias` to the `Perceptron`. 

> A `bias` is a constant that helps the `Perceptrons` input to have a `non-zero` value. 

Bias is a `constant` That we can add to the input to make sure that the input is never `0`. So, the weights can have an effect on the output. We can represent the `bias` as `b`. And the `Perceptron` will look like this:

```mermaid

graph LR
    A[X1*w1] --> |+b1| B(("Perceptron or f(x)"))
    C[X2*w2] --> |+b2| B
    B --> D[y]

```

So, the `Perceptron` will take `input` from `X1` and `X2`, `multiply` it with `weights` `w1` and `w2`, `add` the `bias` `b`, `process` it, and provide `output` as `y`.

$$ y = (w_1x_1 + b_1) + (w_2x_2 + b_2) $$

So, now we can `adjust` the `weights` and `bias` to make the `Perceptrons` more `flexible` even with the `input` as `0`.

We can `generalize` the `Perceptron` as:

$$ y = w_1x_1 + w_2x_2 + b $$

Now think of `n` number of `inputs`. We can represent the `Perceptron` as:

```mermaid

graph LR
    A[X1*w1] --> |+b1| B(("Perceptron or f(x)"))
    C[X2*w2] --> |+b2| B
    D[Xn*wn] --> |+bi| B
    B --> E[y]

```

Well, here's is the `mathematical model` of the `Perceptron` with `n` number of `inputs` and we can finally generalize the `Perceptron` as:

$$ y = w_1x_1 + w_2x_2 + ... + w_nx_n + b $$
$$ y = \sum_{i=1}^{n} w_ix_i + b $$

> $b$ is the total bias of all the inputs.

With this we can have modeled a `biological neuron` as a `simple perceptron`. Later in this article, we will see how we can `expand` the `input` be a `tensor` of information(n-dimensional array) and how we can `expand` the `Perceptron` to a `Neural Network`.


# Artificial Neural Networks

A neural network is a `multi-layer perceptron` model that is designed to `recognize patterns` in the `data`. 



A `Neural Network` is made up of `layers` of `Perceptrons`. You can clearly see that there are `five layers` in the `Neural Network` above. 

- The `first layer` is called the `input layer` and it takes the `input` from the `data`.

- The `middle layers` are called the `hidden layers` and these layers `learn` to `recognize patterns` in the `data`.

- The `last layer` is called the `output layer` and it `provides` the `output` from the `hidden layer`.

Every `Perceptron` in a `layer` is `connected` to every `Perceptron` in the `next layer` and works as a `input` for the `next layer` and so on.

This `multi-layer perceptron` model is called a `Feedforward Neural Network` because the `input` is `fed forward` through the `network` and the `output` is `provided` at the `end`.

Every `Neural Network` consists of `three layers`:

- `Input Layer`
- `Hidden Layer`
- `Output Layer`

> Input Layer: The `input layer` is the `first layer` of the `Neural Network`. It can have `n` number of `neurons/perceptrons` where `n` is the number of `features` in the `data`.

> Output Layer: The `output layer` is the `last layer` of the `Neural Network`. It can have `m` number of `neurons/perceptrons` where `m` is the number of `classes` in the `data` meaning the number of `output` we might have. Also if the output is numerical then the `output layer` is called the `regression layer` and it can only have one `neuron`. 

> `Hidden Layer`: The `hidden layer` is the `middle layer` of the `Neural Network`. A hidden layer can have `n` number of `layers` and it's because of the `hidden layer` that the `Neural Network` can `learn` `complex patterns`. The `hidden layer` is the `magic` of the `Neural Network` where all the `processing` is done and the `patterns` are `learned`.

`Hidden layers` are hard to `interpret` and `understand` because every `Perceptron` in the `hidden layer` takes input from every `Perceptron` in the `previous layer`, so the internal workings become more and more mathematically complicated. As the layers get deeper, the internal workings become more and more mathematically complicated.

> Hidden layers have high connectivity and are distantly connected to the input and output layers. 

> A `Neural Network` becames a `Deep Neural Network` when it has more than one `hidden layer`.

So, the picture above is a `Feedforward Neural Network` with `three layers` and three `hidden layers`. So, we can say that this is a `Deep Neural Network`.

Previously, we have seen that a `Perceptron` can be represented as a `mathematical model` or a `function` like:

$$ y = w_1x_1 + w_2x_2 + ... + w_nx_n + b $$

But this is a linear function and if we think of this as a `Neural Network`, it'll still be a `linear function`. But a dataset with non linear values cannot actually be represented by a `linear function`. 

So, we need to `generalize` the `Perceptrons` to `non-linear` values.

> We can do that with `Activation Functions`.

## Activation Functions

The `W` clearly implies the weight of incoming `input`.

`b` is the `bias` value which is a constant value that helps the `Perceptrons` input to have a `non-zero` value. We can also think of the `bias` as a `offset` value that makes **xw** have to reach a certain threshold before it can have any effect on the output.

For example if `b=-10` in `x*w+b` then the effects of `x*w` won't really start to overcome the bias until `x*w` is greater than `10`.

After that, the effect will solely be determined by the value of `w` or the weight of the input.

That's why it's called a `bias` because it can `bias` the output of the `Perceptron` to be `positive` or `negative` depending on the value of the `bias`.

Now, the `output` of the `Perceptron` is `linear` in nature. It is a `straight line` and it can only `classify` `linearly separable` problems. But in the real world, we have `non-linear` problems that cannot be `classified` by a `linear` `Perceptron`.

So, we need to `introduce non-linearity` to the `output` of the `Perceptron` to `classify` `non-linear` problems. This is where the `Activation Functions` come in.

> To be completely dumb, you can say that Activation Functions are functions that are added to the outputs of the Perceptrons to transform the output into a non-linear form.





There are many `Activation Functions` that are used in `Neural Networks`. Some of the most popular `Activation Functions` are:

- `Sigmoid` (Logistic Function)
- `Tanh` (Hyperbolic Tangent)
- `ReLU` (Rectified Linear Unit)

### Sigmoid Activation Function

The `Sigmoid Activation Function` is a `non-linear` function that is used in `binary classification` problems. It is also called the `Logistic Function`.

The `Sigmoid Activation Function` is represented as:

$$ f(x) = \frac{1}{1 + e^{-x}} $$

The `Sigmoid Activation Function` takes the `output` of the `Perceptron` and `squashes` it between `0` and `1`. This is useful in `binary classification` problems where we need the `output` to be between `0` and `1`. You have done `logistic regression` before, right? It's the same thing.




### Tanh Activation Function

It is also called the `Hyperbolic Tangent`.

The `Tanh Activation Function` is represented as:

$$ f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

The `Tanh Activation Function` takes the `output` of the `Perceptron` and `squashes` it between `-1` and `1`. This is useful in `binary classification` problems where we need the `output` to be between `-1` and `1`.



### ReLU Activation Function

One the most popular and widely used Activation Function is the `ReLU Activation Function`. `ReLU` stands for `Rectified Linear Unit`.



ReLU is a `linear` function that `returns` `0` if the `output` is `less than 0` and `returns` the `output` if the `output` is `greater than 0` as shown in the picture above.

The `ReLU Activation Function` is represented as:

$$ f(x) = max(0, x) $$

This function shows that the `output` will only be `0` if the `output` is `less than 0`.

The `ReLU Activation Function` is used in `Deep Neural Networks` because it is `computationally efficient` and it `speeds up` the `training` of the `Neural Network`. There have been many `improvements` to the `ReLU Activation Function` like the `Leaky ReLU` and the `Parametric ReLU` and there are still research going on to `improve` the `ReLU Activation Function`.

There are many more `Activation Functions` that are used in `Neural Networks`. But these are the most popular `Activation Functions` that are used in `Deep Neural Networks`.

You can visit the `wiki` page of `Activation Functions` to learn more about the `Activation Functions` that are used in `Neural Networks`. (https://en.wikipedia.org/wiki/Activation_function)

## <a id='toc2_4_'></a>[Multi-Class Classification](#toc0_)

This is a `classification` problem where we have `multiple classes` in the `data`. So, in the activation function, we saw how we can solve the problem of `binary classification` with the `Sigmoid`, `Tanh`, and `ReLU` activation functions. But what if we have `multiple classes` in the `data`?

It's farely simple if you think about it. 

So, By multiple classes, I mean that we have `more than two classes` for a `classification` problem. 

There can be two main types of `multi-class situations`:

- Non-Exclusive Classes
- Mutually Exclusive Classes

*** Non-Exclusive Classes

In `non-exclusive classes`, the `classes` are `not mutually exclusive`. This means that a `data point` can belong to `more than one class`. For example, a `data point` can be a `cat` and a `dog` at the same time or a `data point` can be a `car` and of `red color` at the same time.

*** Mutually Exclusive Classes

In `mutually exclusive classes`, the `classes` are `mutually exclusive`. This means that a `data point` can belong to `only one class`. For example, a `data point` can only be a `cat` or a `dog` but not both.

So, in `multi-class classification`, we need to have `multiple output neurons` in the `output layer` of the `Neural Network`. Each `output neuron` will represent a `class` in the `data`.

So, suppose we have `three classes` in the `target`, we must have `three output neurons` in the `output layer` of the `Neural Network`. Each `output neuron` will represent a `class` in the `data`.


So, if the `target class` has `n` number of `classes`, we must have `n` number of `output neurons` in the `output layer` of the `Neural Network`.

The `output` of the `Neural Network` will be a `vector` of `probabilities` where each `element` of the `vector` will represent the `probability` of the `data point` belonging to a `class`.

So, we cannot just have categories like `red`, `blue`, `green` in the `output layer`. We must have `probabilities` like `0.8`, `0.1`, `0.1` where `0.8` is the `probability` of the `data point` belonging to the `red` class and `0.1` is the `probability` of the `data point` belonging to the `blue` and `green` class.

So, in a `multi-class classification` problem, we can use the `one-hot encoding` technique to `encode` the `target` into a `vector` of `probabilities`.

If you don't know what `one-hot encoding` is, Here's a simple explanation:

`One-hot encoding` is a `technique` that is used to `convert` `categorical data` into a `numerical` `vector` of `0s` and `1s`. SO, duppose we have a dataset like this:

| data | target |
|------|--------|
| 1    | red    |
| 2    | blue   |
| 3    | green  |
| 4    | red    |

We can `one-hot encode` the `target` like this:

| data | red | blue | green |
|------|-----|------|-------|
| 1    | 1   | 0    | 0     |
| 2    | 0   | 1    | 0     |
| 3    | 0   | 0    | 1     |
| 4    | 1   | 0    | 0     |

So, `one-hot encoding` breaks the `target` into `multiple columns` where each `column` represents a `class` in the `data` then `encodes` the `target` into a `vector` of `0s` and `1s`.

Now, we have the data `organized` and `ready` for `multi-class classification`. Now, all we have to do is set up the right `activation function` for the output layers.

For `non-exclusive classes`, we can use the `Sigmoid` activation function in the `output layer`.

As I have talked about the `Sigmoid` activation function before, it `squashes` the `output` between `0` and `1` and gives us the `probability` of the `data point` belonging to a `class` in between `0` and `1`.

SO, we can just choose the `class` with the `highest probability` as the `predicted class`.

For `mutually exclusive classes`, we can use a cleaver `activation function` called the `Softmax` activation function in the `output layer`.

The `Softmax` activation function is a `generalization` of the `Sigmoid` activation function. It also `squashes` the `output` between `0` and `1` but it also `normalizes` the `output` to `sum` to `1`. What does this mean?

Suppose we have a `data point` that can belong to `three classes` like `red`, `blue`, and `green`. If we use `sigmoid` activation function, let say the output is `[0.8, 0.3, 0.1]`. We can clearly see that the first `class` has the `highest probability` of the `data point` belonging to the `red` class. But here's the catch, the `output` is not `normalized`, you can get the `output` like `[0.8, 0.8, 0.8]` which is not `normalized` and it's not good for `classification`.

So, the `Softmax` activation function `normalizes` the `output` to `sum` to `1`. This means that whatever happens, the `output` will always `sum` to `1`. So, the `output` will be `[0.5, 0.3, 0.2]` which is `normalized` and it's good for `classification`.

`Softmax` activation function is used in `multi-class classification` problems where the `classes` are `mutually exclusive`. It is represented as:

$$ f(x) = \frac{e^{x_i}}{\sum_{j=1}^{n} e^{x_j}} $$

Here, `n` is the number of `classes` in the `data` and you can see that the `j` is the `index` of the `classes` in the `data`.

So, this gives us a `normalized` `output` because the `output` will always `sum` to `1` .

*** Quadratic Cost Function or Mean Squared Erre(MSE)

The `quadratic cost function`, also known as the `mean squared error (MSE)`, is one of the commonly used `cost functions` in `neural networks`. It measures the average squared difference between the `predicted output` and the `actual output`.

Mathematically, the `quadratic cost function` is defined as:

$$ C = \frac{1}{2n} \sum_{i=1}^{n} (y_i - a_i)^2 $$

Here, `C` represents the `cost`, `n` is the number of `training examples`, `y_i` is the `actual output` for the i-th `training example`, and `a_i` is the `predicted output` for the i-th `training example`.

Why are we `squaring` the difference between the actual and predicted output?

The `quadratic cost function` penalizes larger errors more heavily due to the `squaring operation`. It aims to minimize the overall squared difference between the `predicted` and `actual outputs`, making it suitable for `regression problems`.

> Sometimes there will be some values that are going to be way off the actual value. So, we need to penalize them more heavily. This is why we square the difference between the actual and predicted output.

During the `training process`, the `neural network` adjusts its `weights and biases` using `gradient descent` to minimize the `quadratic cost function`. By iteratively updating the `weights and biases`, the `neural network` learns to make more accurate predictions.

Although the `quadratic cost function` is widely used, it may not be suitable for all types of problems. In some cases, such as `classification problems` with `mutually exclusive classes`, other `cost functions` like `cross-entropy loss` are preferred.

*** Cross-Entropy Loss

The `cross-entropy loss`, also known as `log loss`, is another commonly used `cost function` in `neural networks`. It is particularly well-suited for `classification problems` with `mutually exclusive classes`.

Mathematically, the `cross-entropy loss` is defined as:

$$ C = -\frac{1}{n} \sum_{i=1}^{n} [y_i \log(a_i) + (1 - y_i) \log(1 - a_i)] $$

FOR M Number of classes:

$$ C = -\sum_{j=1}^{M} y_{ij} \log(a_{ij}) $$

Here, `C` represents the `cost`, `n` is the number of `training examples`, `y_i` is the `actual output` for the i-th `training example`, and `a_i` is the `predicted output` for the i-th `training example`.

The `cross-entropy loss` penalizes the `predicted output` based on how well it matches the `actual output`. It aims to minimize the difference between the `predicted` and `actual outputs`, making it suitable for `classification problems`.

During the `training process`, the `neural network` adjusts its `weights and biases` using `gradient descent` to minimize the `cross-entropy loss`. By iteratively updating the `weights and biases`, the `neural network` learns to make more accurate predictions.

The choice of `cost function` depends on the `nature of the problem` and the `type of output` the `neural network` is expected to produce. By selecting an appropriate `cost function`, the `neural network` can be trained to make accurate predictions for a wide range of applications.

The `cost function` is not dependent on a single `weight`. It is dependent on `thousands` of `weights` and `biases`. Which means this curve is not a `2D` curve but a `n-dimensional` curve. So, we cannot just plot out the `cost function` and find the `minimum` value of the `weights` to minimize the `cost function` but we have to take every weight in consideration to minimize the `cost function`.



Since we cannot plot out the `cost function` to find the `minimum` value of the `weights`, we have to use a `mathematical` `algorithm` to `minimize` the `cost function`. This is where the `Gradient Descent` algorithm comes in.

### <a id='toc2_5_2_'></a>[Gradient Descent](#toc0_)

`Gradient Descent` is a `mathematical` `algorithm` that is used to `minimize` the `cost function` of the `Neural Network`. It is an `iterative` `optimization` `algorithm` that `updates` the `weights` and `biases` of the `Neural Network` to `minimize` the `cost function`.

SO, if we go back to that `2D` curve, we can start from any point on the curve and `descend` to the `bottom` of the curve to find the `minimum` value. How do we do that?

You should know some calculus to understand this better.

- We can calculate the `slope` of the `curve` at a `point` on the `curve`. The `slope` of the `curve` at a `point` is the `derivative` of the `curve` at that `point`. The `derivative` of the `curve` at a `point` tells us the `direction` in which the `curve` is `increasing` or `decreasing`.
- If the `slope` is `positive`, we move `left` to `descend` to the `bottom` of the curve.
- If the `slope` is `negative`, we move `right` to `descend` to the `bottom` of the curve.
- This repeats untill eventually the `slope` is `0` and we have reached the `bottom` of the curve.




Now if you paid attention you would have figured out that we are at the `bottom` of the curve and while doing so we also `update` the `weights` and `biases` of the `Neural Network`, So the updated `weights` and `biases` will be the `minimum` value of the `cost function`.

Now, here's a question for you to think about. We started from random `point` on the `curve` and `descended` to the `bottom` of the curve. But how do we choose the next `point` to `descend` to the `bottom` of the curve?

The answer is `learning rate`.

`Learning rate` is a `hyperparameter` that is used to `control` the `step size` of the `gradient descent` `algorithm`. 

> steps size refers to the `distance` we move to `descend` to the `bottom` of the curve.

If the `learning rate` is `small`, we take `small steps` to `descend` to the `bottom` of the curve. Which takes a lot of time to `converge` to the `minimum` value of the `cost function` but the right `minimum` value of the `cost function` is `guaranteed`.

If the `learning rate` is `large`, we take `large steps` to `descend` to the `bottom` of the curve. Which takes less time to `converge` to the `minimum` value of the `cost function` but this might lead to `overshooting` the `minimum` value of the `cost function`. So, we might not get the right `minimum` value of the `cost function`.

So, the `learning rate` is a `hyperparameter` that is used to `control` the `step size` of the `gradient descent` `algorithm`. It is `tuned` to get the right `minimum` value of the `cost function`.

#### <a id='toc2_5_2_1_'></a>[Adaptive gradient descent](#toc0_)

`Adaptive gradient descent` is a `variant` of the `gradient descent` `algorithm` that `adapts` the `learning rate` during the `training process`. It `adjusts` the `learning rate` based on the `gradient` of the `cost function` at each `iteration`.

In the discussion above, we talked about the `fixed learning rate` that is `manually set` before the `training process`. So, the `fixed learning rate` is `constant` throughout the `training process`.

But what we can do is `adapt` the `learning rate` based on the `gradient` of the `cost function` at each `iteration`. This is called `adaptive gradient descent`.

We can start with larger `learning rate` and `slowly decrease` the `learning rate` as we the `gradient` of the `cost function` `decreases` and gets closer to `0`. This will help us to `converge` to the `minimum` value of the `cost function` more `efficiently` and `quickly`.

There are many `adaptive gradient descent` `algorithms` that are used in `Neural Networks`. Some of the most popular `adaptive gradient descent` `algorithms` are:

- `Adagrad` (Adaptive Gradient Algorithm)
- `RMSprop` (Root Mean Square Propagation)
- `ADAM` (Adaptive Moment Estimation)
    - In 2015, Kingma and Ba introduced the `ADAM` `algorithm` which is an `adaptive` `gradient descent` `algorithm` that `adapts` the `learning rate` based on the `first` and `second moments` of the `gradients`.
    - The `ADAM` `algorithm` `combines` the `advantages` of `Adagrad` and `RMSprop` `algorithms` and `performs` well on a wide range of `Neural Networks`.

    

We will use the `ADAM` `algorithm` in our `Neural Network` implementation.


So, we while finding the `minimum` value of the `cost function`, we are actually finding the `descent` in a n-dimensional curve for all the `weights`. And as we are dealing with these n-dimensional curves, the notations change from `slope` to `gradient`. 

This means we calculate,

$$ \nabla C = \frac{\partial C}{\partial w} $$

## <a id='toc2_6_'></a>[Backpropagation](#toc0_)

`Backpropagation` is a very `mathemetically` `intensive` `algorithm` that is used to `train` the `Neural Network`.

If you thoght that `Gradient Descent` was `complicated`, `Backpropagation` is `more complicated` than that with heavy `derivative` and `chain rule` calculations.

I genuinely don't want to go through the `maths` of the `Backpropagation` algorithm because it's `too complicated` and `hard to understand`. But I will try to explain the `Backpropagation` algorithm in a `simple` and `easy` way.

So, let's start with the `Forward Pass`.

### <a id='toc2_6_1_'></a>[Forward Pass](#toc0_)

The `Forward Pass` is the `first step` in the `Backpropagation` `algorithm`. It is the `process` of `feeding` the `input` through the `Neural Network` to `generate` the `output`.

During the `Forward Pass`, the `input` is `fed` through the `input layer` and `propagated` through the `hidden layers` to `generate` the `output` from the `output layer`.

The `output` from the `output layer` is `compared` with the `actual output` to `calculate` the `cost function`. The `cost function` is a `measure` of how `well` the `Neural Network` is `performing`.

The `cost function` is then `used` to `update` the `weights` and `biases` of the `Neural Network` during the `Backward Pass`.

### <a id='toc2_6_2_'></a>[Backward Pass](#toc0_)

The `Backward Pass` is the `second step` in the `Backpropagation` `algorithm`. It is the `process` of `updating` the `weights` and `biases` of the `Neural Network` to `minimize` the `cost function`.

During the `Backward Pass`, the `gradient` of the `cost function` with respect to the `weights` and `biases` is `calculated` using the `chain rule` of `calculus`.

The `chain rule` of `calculus` is used to `calculate` the `partial derivatives` of the `cost function` with respect to the `weights` and `biases` of the `Neural Network`.

The `partial derivatives` are then `used` to `update` the `weights` and `biases` of the `Neural Network` using the `gradient descent` `algorithm`.

The `weights` and `biases` are `updated` in the `opposite direction` of the `gradient` of the `cost function` to `minimize` the `cost function`.

The `Backward Pass` is `repeated` for `multiple iterations` until the `cost function` is `minimized` and the `Neural Network` is `trained` to make `accurate predictions`.
